```python
import os
import random
path = 'D:\\pro_of_program\\practical-training-projects\\masterwork'
text = []
for dir in os.listdir(path):
    book = path + '\\' + dir
    print(book)
    with open(book , 'r' , encoding='utf8') as b:
        for i in b.readlines():
            if len(i) <= 5:
                continue
            text.append(i)
        b.close()
# random.shuffle(text)
with open('D:\\pro_of_program\\practical-training-projects\\total_content.txt' , 'a' , encoding='utf8') as f:
    for i in text:
        f.write(i.replace('\n' , '') + '\n')
    f.close()
```
